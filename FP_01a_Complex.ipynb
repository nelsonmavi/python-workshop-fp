# Complex Numbers

In graphics, physics and even quantum computing complex numbers are used continuosly.<br>
One example below, on how to rotate vectors, without employing trigonometric functions

```python
import matplotlib.pyplot as plt
import cmath

# Original and rotated point using complex rotation
original = 1 + 1j
rotation = cmath.rect(1, cmath.pi / 2)  # 90 degrees CCW
rotated = original * rotation

# Prepare the plot
plt.figure(figsize=(6, 6))
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True)
plt.axis('equal')
plt.xlim(-2, 2)
plt.ylim(-2, 2)

# Plot original and rotated vectors
plt.quiver(0, 0, original.real, original.imag, angles='xy', scale_units='xy', scale=1, color='blue', label='Original (1+1j)')
plt.quiver(0, 0, rotated.real, rotated.imag, angles='xy', scale_units='xy', scale=1, color='red', label='Rotated (90Â° CCW)')

# Labels and legend
plt.title("2D Rotation Using Complex Numbers")
plt.legend()
plt.xlabel("Real Axis")
plt.ylabel("Imaginary Axis")
plt.show()
```
